```python
from mplsoccer import Pitch, VerticalPitch
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2,2)

pitch = Pitch(axis = True, label = True)
pitch.draw(axs[0,0])
axs[0,0].set_xlabel("length")

pitch = Pitch(pitch_color = 'green')
pitch.draw(axs[0,1])
axs[0,1].set_title("Pitch")

pitch = Pitch(axis = True, label = True)
pitch.draw(axs[1,0])

pitch = Pitch(positional = True)
pitch.draw(axs[1,1])
axs[1,1].set_xlabel("Label")

fig.tight_layout()

plt.show()
```

```python
from mplsoccer import Pitch, VerticalPitch
import matplotlib.pyplot as plt
import numpy as np

pitch = Pitch(pitch_color = 'g')
fig, ax = pitch.draw()

x = np.random.randint(0, pitch.dim.right, 11)
y = np.random.randint(0, pitch.dim.bottom, 11)
ax.scatter(x,y ,c='white')

x2 = np.random.randint(0, pitch.dim.right, 11)
y2 = np.random.randint(0, pitch.dim.bottom, 11)
ax.scatter(x2,y2 ,c='yellow')

plt.show()
```
